## Use ydata-profiling to generate a report of the provided dataset. 

### The Tunisian dataset appears long, so we will work with just a random set

```python
#Take a random sample of the determined rate
sample_df = Tunisian_dataset.sample(n =170, random_state =35) 

```

```python
## Generating Profile report
profile = ProfileReport(sample_df, title= 'Pandas Profiling Report')
```

```python
#Generate an HTML report
profile.to_file("sample_df.html")   
 
```

```python
profile.to_notebook_iframe()
```

```python
The report indicates that there are no missing values in the dataset.
```
